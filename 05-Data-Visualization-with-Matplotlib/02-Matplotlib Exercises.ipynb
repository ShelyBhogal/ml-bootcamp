## Exercise 1

**Follow along with these steps:**

* **Create a figure object called 'fig' using `plt.figure()`**
* **Use `.add_axes()` to add an axis to the figure object at [0,0,1,1]. Call this new axis 'ax'.**
* **Plot (x, y) on that axes and set the labels and titles to match the plot below:**

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title');

# Adding semi-colon ; at end of plotting commands removes the output cell
```

## Exercise 2
**Create a figure object and put two axes on it, ax1 and ax2. Located at [0, 0, 1, 1] and [0.2, 0.5, 0.2, 0.2] respectively.**

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
```

**Now plot (x, y) on both axes. And call your figure object to show it.**

```python
ax1.plot(x, y, 'r')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax2.plot(x, y, 'r')
ax2.set_xlabel('x')
ax2.set_ylabel('y')

# Call figure object
fig
```

## Exercise 3

**Create the plot below by adding two axes to a figure object at [0, 0, 1, 1] and [0.2, 0.5, 0.4, 0.4]**

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])
```

**Now use x, y and z arrays to recreate the plot below. Notice the x and y limits on the smaller plot:**

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

ax1.plot(x, z)
ax1.set_xlabel('X')
ax1.set_ylabel('Z')

ax2.plot(x, y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_ylim([30, 50])
ax2.set_xlim([20, 22])
ax2.set_title('zoom-in')

# Note: Without semi-colon terminating plottinng commands, you see output cell (Text(0.5, 1.0, 'zoom-in'))
```

## Exercise 4

**Use `plt.subplots(nrows=1, ncols=2)` to create the plot below.**

```python
fg, axes = plt.subplots(nrows=1, ncols=2)
```

**Now plot (x, y) and (x, z) on the axes. Play around with the linewidth and style.**

```python
fg, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y, color='blue', linewidth=3, linestyle='--')

axes[1].plot(x, z, color='red', linewidth=3)

plt.tight_layout();
```

**See if you can re-size the plot by adding the `figsize=()` argument in plt.subplots() by copying-and-pasting your previous code.**

```python
fg, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 2))

axes[0].plot(x, y, color='blue', linewidth=3, linestyle='--')

axes[1].plot(x, z, color='red', linewidth=3);
```
