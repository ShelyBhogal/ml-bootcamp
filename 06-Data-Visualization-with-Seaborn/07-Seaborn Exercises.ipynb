**Recreate the plots below using the titanic DataFrame. There are very few hints since most of the plots can be done with just one or two lines of code, and a hint would basically give away the solution! Keep careful attention to the *x* and *y* labels for hints.**

***Note! In order to not lose the plot image, make sure you don't code in the cell that is directly above the plot, there is an extra cell above that one which won't overwrite that plot!***

```python
# Joint plot showing relationship between age and ticket fare

sns.jointplot(x='fare', y='age', data=titanic, height=8);
```

```python
# Joint plot showing correlation coefficient

from scipy.stats import pearsonr

# Create JointGrid object
jp = sns.jointplot(data=titanic, x="fare", y="age", xlim=[-100, 600], ylim=[-10, 90], kind="scatter", height=8)

# Calculate correlation coefficient with Stats library
titanic2 = titanic[["fare", "age"]].dropna(axis="rows")
r, p = pearsonr(titanic2["fare"], titanic2["age"])

# 
noplot, = jp.ax_joint.plot([], [], linestyle="", alpha=0)
jp.ax_joint.legend([noplot],[f"r={r:.3f}, p={p:.3f}"]);
```

```python
# Histogram of ticket fare

sns.histplot(data=titanic, x='fare', bins=30, color='red')
plt.xlim([0,600]);
```

```python
# Box plot of Class column distributed over age values

sns.boxplot(x='class', y='age', data=titanic, palette='rainbow');
```

```python
# Swarm plot of class column distributed over age 

plt.figure(figsize=(12, 6))
sns.swarmplot(x="class", y="age", data=titanic, palette='Set2');
```

```python
# Count plot of gender count 

sns.countplot(x='sex', data=titanic)
plt.ylim([0, 600]);
```

```python
# Heat map of numeric variables

plt.figure(figsize=(10, 6))
sns.heatmap(titanic.corr(), annot=True, cbar=True, cmap='coolwarm')
plt.title('Correlation Coefficients');
```

```python
# Using FacetGrid to create individual sub-plots of histograms for age by gender

g = sns.FacetGrid(titanic, col="sex")
g = g.map(plt.hist, "age")
plt.xlim([0, 80])
plt.ylim([0, 120]);
```
