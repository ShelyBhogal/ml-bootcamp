## Style Sheets

Matplotlib has [style sheets](https://matplotlib.org/2.0.2/gallery.html) that you can use to make your plots look a little nicer. These style sheets include `plot_bmh`, `plot_fivethirtyeight`, `plot_ggplot` and more. They basically create a set of style rules that your plots follow. I recommend using them, as they make all your plots have the same look and feel more professional. You can even create your own style sheet if you want your business charts to all have the same professional look (it is a bit tedious though).

Here is how to use them.

**Before `plt.style.use()` your plots look like this:**

```python
df1['A'].hist();
```

Call the style:

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

Now your plots look like this:

```python
df1['A'].hist();
```

```python
plt.style.use('bmh')

df1['A'].hist();
```

```python
plt.style.use('dark_background')

df1['A'].hist();
```

```python
plt.style.use('fivethirtyeight')

df1['A'].hist();
```

```python
plt.style.use('ggplot')
```

Let's stick with the 'ggplot' style and actually show you how to utilize Pandas built-in plotting capabilities!

# Plot Types

There are several plot types built into Pandas, most of them statistical plots by nature:

* df.**plot.area**     
* df.**plot.barh**     
* df.**plot.density**  
* df.**plot.hist**     
* df.**plot.line**     
* df.**plot.scatter**
* df.**plot.bar**      
* df.**plot.box**      
* df.**plot.hexbin**   
* df.**plot.kde**      
* df.**plot.pie**

You can also just call `df.plot(kind='hist')` and replace the `kind` argument with any of the key plotting terms shown in the list above (e.g. 'box','barh', 'scatter', etc.).
___

Let's start going through them!

## Area

```python
# Adjust opacity with alpha parameter
df2.plot.area(alpha=0.4);
```

## Barplots

```python
df2.head()
```

```python
# Re-size figure with figsize parameter
df2.plot.bar(figsize=(10, 7));
```

```python
df2.plot.bar(stacked=True);
```

```python
df2[['a', 'b']].plot.bar();
```

## Histograms

```python
df1.head()
```

```python
df1['A'].plot.hist(bins=50);
```

## Line Plots

```python
# Automatically uses index as x-axis if not specified, e.g. if your index is a timeline
df1.plot.line(y='B', figsize=(12, 3), lw=1);
```

```python
# Plot A vs B...no discernible pattern!
df1.plot.line(x='A', y='D', figsize=(12, 3), lw=1);
```

## Scatter Plots

```python
# To view relationship between columns A and B
df1.plot.scatter(x='A', y='B');
```

You can use `c` argument to color your plot based off another column value (similar to `hue` parameter used in Seaborn).

Use `cmap` to name the colormap to use. 

For all the colormaps, check out: http://matplotlib.org/users/colormaps.html

```python
df1.plot.scatter(x='A', y='B', c='C', figsize=(10, 6), cmap='coolwarm');
```

Use `s` to indicate size of dots based off another column (instead of using colour). 

The `s` parameter must be an array, not the name of the column:

```python
df1.plot.scatter(x='A', y='B', s=df1['C'] * 200, figsize=(10, 6))
plt.xlim([-3, 4]);
```

```python
# Use both at the same time!
df1.plot.scatter(x='A', y='B', s=df1['C'] * 200, c='C', figsize=(10, 6), cmap='seismic')
plt.xlim([-3, 4]);
```

## Hexagonal Bin Plot

Useful for Bivariate Data, alternative to scatterplot:

```python
# Create 1000x2 DataFrame of values randomly-selected from 'normal' distribution
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])

# Gridsize increases number of hexagons in x-axis direction
df.plot.hexbin(x='a', y='b', gridsize=25, cmap='Oranges');
```
